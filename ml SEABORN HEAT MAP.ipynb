```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data=np.random.randint(low=1,
                       high=100,
                       size=(10,10))
#setting the parameter value

vmax=20
vmin=80

hm=sns.heatmap(data=data,
               vmin=vmin,
               vmax=vmax)

plt.show()
               
                       
```
