```python
# matplotlib markers

import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([3,8,1,10])

plt.plot(ypoints,marker='o')
plt.show()



```

```python
# marker/line/color/
# .....................................

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([4,8,6,12])
plt.plot(ypoints,'o:r')
plt.show()
```

```python
# marker size
# ------------------------
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([2,8,6,7,25])
plt.plot(ypoints,'o:r',ms=35)
plt.show()
```

```python
# marker color
# ...........................


import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array ([5,9,12,6,7])
plt.plot(ypoints,marker='o',ms=20,mec='r')
plt.show()
                  
```

```python
# ADDING GRID LINES
# ---------------------------------

import numpy as  np
import matplotlib.pyplot as plt

x= np.array([80,85,90,95,100,105,110,115,120,125])
y= np.array([240,250,260,270,280,290,300,310,320,330])
plt.title ("sports watch data")
plt.xlabel("Average pulse")
plt.ylabel("colorie burnage")
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
# SPECIFY WHICH GRID LINES TO DISPLAY
# -----------------------------------------

import numpy as np
import matplotlib.pyplot as plt
x=np.array([80,85,90,95,100,110,115,120])
y=np.array([240,250,260,270,280,290,300,310])
plt.title("sports watch data")
plt.xlabel("average pulse")
plt.ylabel("calorie burnage")
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
# DISPLAY ONLY GRID LINES FOR THE X AXIS
# .........................................

import numpy as np
import matplotlib.pyplot as plt
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.title("sports watch data")
plt.title("average pulse")
plt.ylabel("calorie burnage")
plt.plot(x,y)
plt.grid(axis='x')
plt.show()
```
